### Carregando a tabela com os dados da m√°quina e visualizando

```python
pdm_machine_raw = pd.read_csv('C:/Users/Joelson Silva/VS Code/Banco de Dados I/Projeto 1/CSVs/PdM_machine.csv')
pdm_machine_raw.head()
```

```python
model=pdm_machine_raw['model'].unique()
model.sort()
print(model)
pdm_model_machine = pd.DataFrame(columns=['modelID','model_machine'])
pdm_model_machine['model_machine']=model
pdm_model_machine['modelID']=[x+1 for x in range(len(pdm_model_machine['modelID']))]
pdm_model_machine
```

```python
pdm_model_machine.to_sql('PdM_model_machine', con=db, if_exists='append', index=False)
sql = """
    SELECT * 
    FROM PdM_model_machine
"""
model_machine = pd.read_sql(sql=sql, con=db)
model_machine.head()
```

```python
pdm_machine = pdm_machine_raw.rename(columns={'model':'modelID'})
pdm_machine['modelID']=pdm_machine['modelID'].apply(lambda x: int(x[-1]))
pdm_machine.head()
```

```python
pdm_machine.to_sql('PdM_machine', con=db, if_exists='append', index=False)
sql = """
    SELECT * 
    FROM PdM_machine
"""
machine = pd.read_sql(sql=sql, con=db)
machine.head()
```
