Um dia o Prof. Humberto Jos√© Roberto fez o seguinte questionamento: Se o zero a esquerda de um n√∫mero n√£o tem valor algum, por que teria em outras posi√ß√µes de um n√∫mero? Analisando da seguinte forma, ele pede sua ajuda para, ao somar dois valores inteiros, que o resultado seja exibido segundo o racioc√≠nio dele, ou seja, sem os Zeros. Por exemplo, ao somar 15 + 5, o resultado seria 20, mas com esta nova ideia, o novo resultado seria 2, e, ao somar 99 + 6, o resultado seria 105, mas com esta nova ideia, o novo resultado seria 15.

Escreva um programa que, dado dois n√∫meros inteiros, sem o algarismo zero, some os mesmos e, caso o resultado tenha algum algarismo zero, que os retire antes de exibir.

Entrada

Haver√° diversos casos de teste. Cada caso de teste inicia com dois inteiros M e N (1 ‚â§ M ‚â§ N ‚â§ 999.999.999).

O √∫ltimo caso de teste √© indicado quando N = M = 0, sendo que este caso n√£o deve ser processado.

Sa√≠da

Para cada caso de teste, imprima o resultado da soma dos dois valores, sem os Zeros.

```python
def sumHumberto(M, N):
  resultado = M + N
  resultado_string = list(str(resultado))
  for indice, letra in enumerate(resultado_string):
    if letra == "0":
      resultado_string.pop(indice)
  return int("".join(resultado_string))

sumHumberto(M = 99,N = 6)
```

**Objetivo:** saber trabalhar com strings √© uma habilidade muito importante no dia a dia dos profissionais de dados. Por isso:

- Crie uma fun√ß√£o que seja capaz de contabilizar a ocorr√™ncia de uma mesma letra dentro de uma string; e
- Crie uma fun√ß√£o que seja capaz de contabilizar a ocorr√™ncia de uma mesma palavra dentro de uma string.
Aten√ß√£o: Utilize os conceitos vistos em aula para gerar sua solu√ß√£o.

```python
def contaLetras(letra: str, palavra: str):
  contador = 0
  for i in palavra.strip(): 
    if letra == i: 
      contador += 1
  if contador == 0: 
    print("Essa letra n√£o consta na palavra buscada!")
  else:
    print(f"A quantidade de vezes que essa letra se repete √© {contador}.")
contaLetras('a','Gustavo')
```
