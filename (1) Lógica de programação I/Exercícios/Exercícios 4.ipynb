# Crie um programa que compara duas strings e retorna o caractere incorreto.

```python
def comparaString(nome = 'casa', nome2 = 'caza'): 
    caracteres_invalidos = []
    for i in range(len(nome)):
            if(nome[i] != nome2[i]):
                caracteres_invalidos.append(nome[i])
    print(f'Os caracteres que estão diferentes são {caracteres_invalidos}')
comparaString()
```

# Faça um programa que conte o numero de 1’s que aparecem em um string. Exemplo: ´ 0011001 -> 3

```python
sequencia_n = input("Digite uma sequência de números")
contador = 0
qtde = 0

while(contador < len(sequencia_n)): 
    if(sequencia_n[contador] == '1'):
        qtde +=1
    contador +=1
if(qtde == 1):
    print(f'A quantidade de vezes que o número um apareceu na sequência digitada foi {qtde} vez') 
else: 
    print(f'A quantidade de vezes que o número um apareceu na sequência digitada foi {qtde} vezes') 
```

# Escreva um programa que substitui as ocorrencias de um caractere 0 em uma string por outro caractere 1.

```python
sequencia_n = input("Digite uma sequência de números")
lista = list(sequencia_n)
contador = 0

while(contador < len(sequencia_n)): 
    if(lista[contador] == '0'):
        lista[contador] = '1'
    contador +=1
sequencia_n = "".join(lista) # join é utlizado para unir os slementos de uma lista em uma única string, podendo ser separado 
# por algum caractere, exemplo lista = [1,2,3] "-"join.Lista = "1-2-3"
print(sequencia_n)
```

# Faca um programa que receba uma palavra e calcule quantas vogais (a, e, i, o, u) possui essa palavra. 
# Entre com um caractere (vogal ou consoante) e substitua todas as vogais da palavra dada por esse caractere.

```python
palavra = input("Digite a palavra desejada").lower()
caractere = input("Digite o caractere para substituir a palavra").lower()
contador = 0

verificador = list(palavra)
tamanho = len(verificador)

while(contador < tamanho):
    if(verificador[contador] == 'a' or verificador[contador] == 'e' or verificador[contador] == 'i' or verificador[contador] == 'o' or verificador[contador] == 'u'):
        verificador[contador] = caractere
    contador += 1
palavra = "".join(verificador)
print(palavra)
```

# Escreva um programa para converter uma cadeia de caracteres de letras maiusculas em letras minusculas

```python
def converteMaiuscula(): 
    palavra = input("Digite uma palavra")
    print(palavra.swapcase())
converteMaiuscula()
```

# Faça uma função que informe a quantidade de dígitos de um determinado número inteiro informado.

```python
def somaImposto(taxaImposto, custo): 
    percentual = (taxaImposto/100)*custo
    custo += percentual
    print(f'O valor final com os impostos é de R${custo}')
somaImposto(10,1000)
```

```python
# Reverso do número. Faça uma função que retorne o reverso de um número inteiro informado. Por exemplo: 127 -> 721.

def numeroReverso(numero):
    if(type(numero)) != int:
        print("o número precisa ser inteiro!")
    else: 
        reverso = str(numero)
        reverso = reverso[::-1]
        print(reverso)
numeroReverso(127)
```

# Data com mês por extenso. Construa uma função que receba uma data no formato DD/MM/AAAA e devolva uma string no formato D de mesPorExtenso de AAAA.

```python
def data():
    data = input("Digite uma data no formado dd/mm/aaaa").lstrip()
    mes = int(data[3:5])

    if(mes == 1): 
        print(f'{data[0:2]} de janeiro de {data[6:]}')
    elif(mes == 2):
        print(f'{data[0:2]} de fevereiro de {data[6:]}')
    elif(mes == 3): 
        print(f'{data[0:2]} de março de {data[6:]}')
    elif(mes == 4): 
        print(f'{data[0:2]} de abril de {data[6:]}')
    elif(mes == 5): 
        print(f'{data[0:2]} de maio de {data[6:]}')
    elif(mes == 6): 
        print(f'{data[0:2]} de junho de {data[6:]}')
    elif(mes == 7): 
        print(f'{data[0:2]} de julho de {data[6:]}')
    elif(mes == 8): 
        print(f'{data[0:2]} de agosto de {data[6:]}')
    elif(mes == 9): 
        print(f'{data[0:2]} de setembro de {data[6:]}')
    elif(mes == 10): 
        print(f'{data[0:2]} de outubro de {data[6:]}')
    elif(mes == 11):
        print(f'{data[0:2]} de novembro de {data[6:]}')
    else: 
        print(f'{data[0:2]} de dezembro de {data[6:]}')
data()
```

# Embaralha palavra. Construa uma função que receba uma string como parâmetro e devolva outra string com os carateres embaralhados. Por exemplo: se função receber a palavra python, pode retornar npthyo, ophtyn ou qualquer outra combinação possível, de forma aleatória. Padronize em sua função que todos os caracteres serão devolvidos em caixa alta ou caixa baixa, independentemente de como foram digitados.

```python
import random

def embaralhaPalavra(palavra:str):
  palavra_embaralhada = []
  lista = list(palavra)
  for i in range(len(lista)):
      sorteio = random.randint(0, len(lista)-1)
      palavra_embaralhada.append(lista[sorteio])
      lista.pop(sorteio)
  palavra_embaralhada = "".join(palavra_embaralhada)
  print(palavra_embaralhada)
embaralhaPalavra('teste')
```

# Faça um programa para imprimir:

#     1
#     2   2
#     3   3   3
#     .....
#     n   n   n   n   n   n  ... n
## para um n informado pelo usuário. Use uma função que receba um valor n inteiro e imprima até a n-ésima linha.

```python
def imprimeNumero(numero): 
    lista = list(range(1,numero+1,1))

    for i in range(len(lista)):
      print(lista[i] * f' {i+1} '.center(3))
imprimeNumero(12)
```
