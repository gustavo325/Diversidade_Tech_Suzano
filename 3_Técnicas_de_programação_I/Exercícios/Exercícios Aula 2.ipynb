Considere que apenas entradas únicas devem ser armazenadas na tabela. Caso o registro seja repetido, sinalize a repetição, ID e o resto de seu conteúdo.

```python
import json
import requests
import pandas as pd
```

```python
def cria_dataframe(valor):
  dados = pd.DataFrame(valor.json().items())
  dados = dados.set_index(0)
  dados = dados.T
  return dados
```

```python
def tabela(iteracoes): 
  tabela_final = cria_dataframe(requests.get('https://official-joke-api.appspot.com/random_joke'))
  while iteracoes > 1: 
    dado = cria_dataframe(requests.get('https://official-joke-api.appspot.com/random_joke'))
    tabela_final.append(dado)
    iteracoes -= 1
  return tabela_final
```
