- A quantidade de vezes em que os dois alunos tiraram a mesma nota final
- A média anual dos alunos.
- A quantidade de bimestres que o aluno A tirou melhor nota que B.
- O ano e bimestre que o aluno A teve a maior diferença absoluta em relação a B.

nota_aluno_A = np.array([7, 7.5, 6, 8.8, 9, 10, 9.5, 8]) onde cada valor representa uma nota em um bimestre. Os quatro primeiros valores representam a nota no primeiro ano.

O cálculo do Índice de Massa Corpórea (IMC) de cada pessoa.
A quantidade e proporção de clientes em cada grupo, de acordo com o IMC.
A quantidade de pessoas acima de 90 kg e abaixo de 1.70 m.

```python
dados_cliente = np.array([(1.94, 80), (1.65, 92), (1.81, 86), (1.68, 91)]) # Altura e peso
imc = dados_cliente[:,1:2]/(dados_cliente[:,:1]**2)
classificacao_imc = []
acima_90 = len(dados_cliente[dados_cliente > 90])
altura_baixa = len(dados_cliente[dados_cliente < 1.70])

for valor in imc: 
  if valor[0] < 18.5: 
    classificacao_imc.append('Abaixo do peso')
  elif valor[0] < 24.9: 
    classificacao_imc.append('Peso normal')
  elif valor[0] < 29.9: 
    classificacao_imc.append('Excesso de peso')
  elif valor[0] < 34.9: 
    classificacao_imc.append('Obesidade classe 1')
  elif valor[0] < 39.9: 
    classificacao_imc.append('Obesidade classe 2')
  else: 
    classificacao_imc.append('Obesidade classe 3')

print(f'O IMC de todos os clientes da academia é de\n {imc}.'\
      f'\nA quantidade de clientes acima de 90Kg é {acima_90}.'\
      f'\nA quantidade de clientes abaixo de 1.70 é de {altura_baixa}'\
      f'\nA proporção de clientes abaixo do peso é de {classificacao_imc.count("Abaixo do peso")}/{len(dados_cliente)}.'\
      f'\nA proporção de clientes com peso normal é de {classificacao_imc.count("Peso normal")}/{len(dados_cliente)}.'\
      f'\nA proporção de clientes com excesso de peso é de {classificacao_imc.count("Excesso de peso")}/{len(dados_cliente)}.'\
      f'\nA proporção de clientes com obesidade classe 1 é de {classificacao_imc.count("Obesidade classe 1")}/{len(dados_cliente)}.'\
      f'\nA proporção de clientes com obesidade classe 2 é de {classificacao_imc.count("Obesidade classe 2")}/{len(dados_cliente)}.'\
      f'\nA proporção de clientes com obesidade classe 3 é de {classificacao_imc.count("Obesidade classe 3")}/{len(dados_cliente)}.')
```
