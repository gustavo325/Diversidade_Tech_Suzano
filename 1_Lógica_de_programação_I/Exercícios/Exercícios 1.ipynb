# Escreva um programa que recebe uma altura e retorna se esta é alta, baixa ou média.

```python
altura = float(input("Digite a altura "))

if(altura <= 1): 
    print(f'A altura digitada foi {altura}, e é considerada baixa')
elif(altura <= 3):
    print(f'A altura digitada foi {altura}, e é considerada média')
else: 
    print(f'A altura digitada foi {altura}, e é considerada alta')
```

# Crie um cadastro que recebe nome, telefone, e peça para confirmar as informações ao final.

```python
lista = []

nome = input("Digite um nome")
lista.append(nome)
telefone = int(input("Digite um telefone (somente números)"))
lista.append(telefone)

print(f'Você digitou as seguintes informações, {lista} ')

pergunta = input("As informações estão corretas? Digite (s/n)")
while(pergunta == 'n'):
    pergunta2 = int(input("Qual informação deseja alterar, digite o número correspondente\n1-Nome\n2-telefone"))
    valor = pergunta2 - 1
    if(valor == 1):
        lista[valor] = int(input("Digite um telefone (somente números)")) 
    elif(valor == 0):
        lista[valor] = input("Digite um nome")
    print(f'Você digitou as seguintes informações, {lista} ')
    pergunta = input("As informações estão corretas? Digite (s/n)")
print(f'A resposta final = {lista}')
    
```

# Crie um programa que recebe sua altura e peso, calcula o imc e retorna um status (abaixo do peso, no peso, acima do peso).

```python
altura = float(input("Digite a sua altura... "))
peso = float(input("Digite seu peso... "))

imc = peso/(altura**2)
imc = round(imc, 2)

if(imc < 16):
    print(f'Seu IMC = {imc} e você está com subpeso severo.')
elif(imc < 19.9): 
    print(f'Seu IMC = {imc} e você está com subpeso.')
elif(imc < 24.9): 
    print(f'Seu IMC = {imc} e você está com peso normal.')
elif(imc < 29.9):
    print(f'Seu IMC = {imc} e você está com sobrepeso.')
elif(imc < 39.9): 
    print(f'Seu IMC = {imc} e você está Obeso.')
else: 
    print(f'Seu IMC = {imc} e você está no estado de obesidade morbida... Se cuide!')
```
