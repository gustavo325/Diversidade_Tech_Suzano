# Crie um programa que faça as seguintes 5 perguntas e diagnostique o paciente:

### A. Sente dor no corpo?

### B. Tem febre?

### C. Tem tosse?

### D. Está com congestão nasal?

### E. Tem manchas no corpo?

### Condições:

### Se apenas a primeira for verdade, o paciente não está doente; (10)
### Se B, C e D forem verdade, o paciente está com gripe; (6)
### Se todas forem verdade, paciente está com Dengue; (17)
### Caso contrário, não temos o diagnóstico. 

```python
contador = 0

pergunta1 = input('Sente dor no corpo? (s/n)').upper()
pergunta2 = input('Tem febre? (s/n)').upper()
pergunta3 = input('Tem tosse? (s/n)').upper()
pergunta4 = input('Está com congestão nasal? (s/n)').upper()
pergunta5 = input('Tem manchas no corpo? (s/n)').upper()

if(pergunta1 == 'S'):
    contador += 10
if(pergunta2 == 'S'): 
    contador += 2
if(pergunta3 == 'S'): 
    contador += 2 
if(pergunta4 == 'S'): 
    contador += 2
if(pergunta5 == 'S'): 
    contador += 1

if(contador == 10):
    print("Paciente não está doente!")
if(contador == 6): 
    print("Paciente está com gripe!")
if(contador == 17): 
    print("Paciente está com Dengue!")

if(contador != 10 and contador != 3 and contador != 17):
    print("Não temos o diagnóstico!")
```

# Crie um programa que recebe nome e telefone do usuario. Ele só para quando o usuário pedir para sair.

```python
banco = []

resposta = input("Deseja inserir uma informação ? (s/n)").upper()

while (resposta == 'S'): 
    questao1 = input("Digite o nome do usuário") 
    banco.append(questao1)
    questao2 = input("Digite o telefone do usuário")
    banco.append(questao2)
    resposta = input("Deseja inserir uma outra informação ? (s/n)").upper()
print(banco)
```

# Faça um programa que faça 5 perguntas para uma pessoa sobre um crime. As perguntas são:

### "Telefonou para a vítima?"

### "Esteve no local do crime?"

### "Mora perto da vítima?"

### "Devia para a vítima?"

### "Já trabalhou com a vítima?" 

### O programa deve no final emitir uma classificação sobre a participação da pessoa no crime. 
### Se a pessoa responder positivamente a 2 questões ela deve ser classificada como "Suspeita", 
### entre 3 e 4 como "Cúmplice" e 5 como "Assassino". Caso contrário, ele será classificado como "Inocente".

```python
contador = 0

pergunta1 = input("Telefonou para a vítima? (s/n)").upper()
if(pergunta1 == 'S'):
    contador +=1
pergunta2 = input("Esteve no local do crime? (s/n)").upper()
if(pergunta2 == 'S'):
    contador +=1
pergunta3 = input("Mora perto da vítima? (s/n)").upper()
if(pergunta3 == 'S'):
    contador +=1
pergunta4 = input("Devia para a vítima? (s/n)").upper()
if(pergunta4 == 'S'):
    contador +=1
pergunta5 = input("Já trabalhou com a vítima? (s/n)").upper()
if(pergunta5 == 'S'):
    contador +=1
    
if(contador == 2):
    resultado = 'Suspeito'
elif(contador > 2 and contador <= 4):
    resultado = 'Cúmplice'
elif(contador == 5):
    resultado = 'Assassino'
else: 
    resultado = 'inocente'
    
print(f'O seu veredito é: {resultado}')
```

# Faça um programa que leia um nome de usuário e a sua senha e não aceite a senha igual ao nome do usuário, mostrando uma mensagem de erro e voltando a pedir as informações.

```python
usuario = input("Digite um nome de usuário")
senha = input("Digite uma senha")

while(usuario == senha): 
    print("Usuário não pode ser igual a senha!\n")
    usuario = input("Digite um nome de usuário")
    senha = input("Digite uma senha")
print(f'O usuário digitado foi {usuario} e a senha foi {senha}')
    
```

# Faça um programa que leia e valide as seguintes informações:

### Nome: maior que 3 caracteres;

### Idade: entre 0 e 150;

### Salário: maior que zero;

### Sexo: 'f' ou 'm';

### Estado Civil: 's', 'c', 'v', 'd';

```python
nome = input("Digite o seu nome ")
while(len(nome) < 3): 
    nome = input("Digite o seu nome (favor inserir mais de 3 caracteres) ")
idade = int(input("Digite a sua idade"))
while(idade <=0 or idade > 150): 
    if(idade > 150): 
        idade = int(input("Você é descendente de matusalém??? A idade tem que ser menor que 150 ;)..."))
    else:
        idade = int(input("Nossa cara, você não nasceu ainda??? A idade tem que ser maior que 0 :D ..."))
salario = int(input("Digite o seu salário"))
while(salario <=0): 
    salario = int(input("O tempo de escravidão acabou... Digite um salário maior que 0 :D"))
sexo = input("Digite o seu sexo\n f para feminino ou m para masculino").upper()
while(sexo != 'M' and sexo != 'F'):
    sexo = input('As opções são F para feminino ou M para masculino')
if(sexo == 'M'): 
    sexo = 'Masculino'
else: 
    sexo = 'Feminino'
estado_civil = input("Digite um estado civil\ns para solteiro\nc para casado\nv para viúvo\nd para divorciado").upper()
while(estado_civil != 'S' and estado_civil != 'C' and estado_civil != 'V' and estado_civil != 'D'): 
    estado_civil = input("Opção inválida, digite novamente!!!")
if(estado_civil == 'S'):
    estado_civil = 'Solteiro'
elif(estado_civil == 'C'):
    estado_civil = 'Casado'
elif(estado_civil == 'V'): 
    estado_civil = 'Viúvo'
else: 
    estado_civil = 'Divorciado'
print(f'Seu nome é {nome}, com idade de {idade}, salário de {salario}, do sexo {sexo} e estado civil {estado_civil}')
```
